```python
import numpy as np
```

```python
# Naive version
r11 = lambda n: 1 if n == 0 else r12(n-1)*0.2 + r11(n-1)*0.5
r12 = lambda n: 1 - r11(n)
r21 = lambda n: (n-1)*0.5 + (n-1)*0.2
r22 = lambda n: 1 - r21(n)
```

```python
# Dynamic programming version
def r1x(n):
    r11_db = np.zeros(n)
    for i in np.arange(1,n):
        r11_db[i] = r11_db[i-1]*0.5 + (1 - r11_db[i-1])*0.2
    return r11_db, 1 - r11_db

def r2x(n):
    r21_db = np.zeros(n)
    for i in np.arange(1,n):
        r21_db[i] = r21_db[i-1]*0.5 + (1 - r21_db[i-1])*0.2
    return r21_db, 1 - r21_db
```

```python
r1x(1000)
```
